```python
#Tìm số huy chương của 1 quốc gia
#Tìm số huy chương vàng, bạc, đồng theo màn thi cá nhân

import pandas as pd
url = 'http://winterolympicsmedals.com/medals.csv'
data = pd.read_csv(url)
data
```

```python
data=data.set_index(data['NOC'])
data.loc['CAN']
```

```python
dataframe = data[(data['NOC']=='CAN')&(data['Medal'])].count()
print('Số huy chương Canada đạt được: ',dataframe['Medal'])
```

```python
dataframe = data[(data['NOC']=='CAN')&(data['Medal']=='Gold')].count()
print('Số huy chương vàng Canada đạt được: ',dataframe['Medal'])
```

```python
dataframe = data[(data['NOC']=='CAN')&(data['Medal']=='Silver')].count()
print('Số huy chương bạc Canada đạt được: ',dataframe['Medal'])
```

```python
dataframe = data[(data['NOC']=='CAN')&(data['Medal']=='Bronze')].count()
print('Số huy chương đồng Canada đạt được: ',dataframe['Medal'])
```

```python
#Môn thi Skiing
dataframe = data[(data['NOC']=='CAN')&(data['Sport']=='Skiing')].count()
print('Số huy chương môn Skating Canada đạt được: ',dataframe['Medal'])
```

```python
dataframe = data[(data['NOC']=='USA')].count()
print('Số huy chương trung bình: ',dataframe['Medal'])
```
